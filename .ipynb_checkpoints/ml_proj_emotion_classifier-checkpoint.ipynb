```python
# this file is used to create the filtered dataset to remove features on the face
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import shutil
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Path to your dataset directory
current_path = os.getcwd()

data_dir = os.path.join(current_path,'fer2013')


# Define transformations for the images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to 224x224
    transforms.ToTensor(),          # Convert images to PyTorch tensors
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])  # Normalize to [-1, 1]
])

# Load the dataset using ImageFolder
train_dataset = datasets.ImageFolder(root=data_dir, transform=transform)

# Create a DataLoader for batching and shuffling
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Print class-to-index mapping
print("Class-to-Index Mapping:", train_dataset.class_to_idx)

# Example: Iterate through the DataLoader
for images, labels in train_loader:
    print("Batch of images shape:", images.shape)  # Shape: [batch_size, channels, height, width]
    print("Batch of labels:", labels)  # Labels corresponding to the classes
    break

```

```python
# plot the distribution of images
num_of_images = []
name_of_classes=[]

# Iterate over subfolders (classes)
for a_class in os.listdir(data_dir + '/train'):
    class_path = os.path.join(data_dir, 'train', a_class)
    if os.path.isdir(class_path):  # Check if it's a directory
        num_of_images.append(len(os.listdir(class_path)))
        name_of_classes.append(a_class)
plt.figure()
plt.bar(name_of_classes,num_of_images,color = "purple")
plt.title("Distribution of different emotions")
plt.xlabel("class_names")
```

```python
# make the test sample smaller, select only 100 pcitures from every class

for a_class in os.listdir(data_dir + '/test'):
    class_path = os.path.join(data_dir, 'test', a_class)
    if os.path.isdir(class_path):  # Check if it's a directory
        output_path = os.path.join(current_path, 'selected_test', a_class)
        # List all images in the class folder
        images = os.listdir(class_path)
        # Select only the first 100 images
        selected_images = images[:100]
        # Copy selected images to the output folder
        for image in selected_images:
            src_path = os.path.join(class_path, image)  # Source file path
            dest_path = os.path.join(output_path, image)  # Destination file path
            shutil.copy(src_path, dest_path)  # Copy the file
        print(f"Copied {len(selected_images)} images for class '{a_class}' to '{output_path}'.")
        print("All images copied successfully.")
    
```

```python
# remove eyes on the picture given an file path
def remove_eye(path,radius=2):
    image = face_recognition.load_image_file(path)
    face_landmarks_list = face_recognition.face_landmarks(image)
    average_color = np.mean(image, axis=(0, 1))

    for face_landmarks in face_landmarks_list:
        for feature, points in face_landmarks.items():
            if feature == "left_eye" or feature == "right_eye":
                for point in points:
                    x, y = point
                    cv2.circle(image, (x, y), dot_radius, average_color, -1)  # Draw white circles
    return image
```

```python
# remove eyes on the picture given an file path
def remove_mouth(path,radius=1):
    image = face_recognition.load_image_file(path)
    face_landmarks_list = face_recognition.face_landmarks(image)
    average_color = np.mean(image, axis=(0, 1))

    for face_landmarks in face_landmarks_list:
        for feature, points in face_landmarks.items():
            if feature == "top_lip" or feature == "bottom_lip":
                for point in points:
                    x, y = point
                    cv2.circle(image, (x, y), dot_radius, (average_color), -1)  # Draw circles
    return image
```
